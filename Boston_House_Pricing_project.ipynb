# Robust Regression

## RANSACRegressor

```python
from sklearn.linear_model import RANSACRegressor
```

```python
df.head()
```

```python
X=df['LSTAT'].values.reshape(-1,1)
```

```python
y=df['MEDV'].values
```

```python
model=RANSACRegressor()
```

```python
model.fit(X,y)
```

```python
line_X=np.arange(1,35,1)
```

```python
line_y=model.predict(line_X.reshape(-1,1))
```

```python
inlier_mask=model.inlier_mask_
outlier_mask=np.logical_not(inlier_mask)
```

```python
plt.figure(figsize=(12,10))
sns.set(style='darkgrid',context='notebook')
plt.plot(line_X,line_y,c='red')
plt.scatter(X[inlier_mask],y[inlier_mask],c='blue',marker='o',label='inlier')
plt.scatter(X[outlier_mask],y[outlier_mask],c='brown',marker='s',label='outlier')
plt.legend(loc='upper right')
plt.xlabel('% lower status of the population')
plt.ylabel('Median value of owner-occupied homes in $1000\'s')
plt.show()
```

```python
model.estimator_.coef_
```

```python
model.estimator_.intercept_
```

# Model Evaluation

```python
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=0)
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
y_test_pred=model.predict(X_test)
y_train_pred=model.predict(X_train)
```

# Method 1 residual method

```python
plt.figure(figsize=(12,10))
plt.scatter(y_test_pred,y_test_pred-y_test,marker='o',c='blue',label='test data')
plt.scatter(y_train_pred,y_train_pred-y_train,marker='*',c='orange',label='training data')
plt.xlabel('predictred values')
plt.ylabel('residuals')
plt.legend(loc='upper left')
plt.hlines(0,1,50,lw=2,color='k')
```

# Method 2 Mean squred method

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_train,y_train_pred)
```

```python
mean_squared_error(y_test,y_test_pred)
```

# Method 3 R squared method

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_train,y_train_pred)
```

```python
r2_score(y_test,y_test_pred)
```
